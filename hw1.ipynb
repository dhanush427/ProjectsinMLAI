### a. Logistic Regression Model
The probability of the binary outcome y âˆˆ {0, 1} can be modeled as:
P(y = 1 | x; Î¸) = h_Î¸(x) = 1 / (1 + exp(-xáµ€Î¸))
P(y = 0 | x; Î¸) = 1 - h_Î¸(x)

### b. Looking at the Likelihood Function
For n independent observations {(xáµ¢, yáµ¢)} (i=1 to n) this is the likelihood function:
L(Î¸) = Î  [h_Î¸(xáµ¢)^(yáµ¢) * (1 - h_Î¸(xáµ¢))^(1 - yáµ¢)]

### c. Log-Likelihood Function
Take the natural log of the likelihood function:
â„“(Î¸) = ln L(Î¸) = Î£ [yáµ¢ * ln(h_Î¸(xáµ¢)) + (1 - yáµ¢) * ln(1 - h_Î¸(xáµ¢))]

### d. Objective Function

Maximum A Posteriori (MAP):

###Why Logistic Regression is the Best Choice
Binary Classification: Logistic regression is specifically designed for binary classification tasks, making it the valid choice for spam detection where the output is binary (ð‘¦âˆˆ{0,1}).

Probabilistic Output: Logistic regression provides the probability of something being in a certain class, which can be useful for threshold-based decision-making and confidence estimation.

Simplicity and Interpretability: The model is simple to implement, computationally efficient, and the models coefficients can help see which features matter the most.

Compare to SVM: The idea is that SVM differs from logistic regression in its aim and methodology. SVM maximizes the margin between classes, focusing only on the more challenging points at the margin, known as support vectors. This approach makes SVM generally robust to outliers but less transparent compared to logistic regression. Logistic regression is computational efficient and time complexity improves much better with large datasets, whereas SVM optimizes in a quadratic manner that can be computationally expensive. Moreover, because logistic regression optimizes for probability variables, it is a better fit for applications demanding confidence score, whereas SVM outputs class labels or decision values that needs further calibration for probabilities.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]

data = pd.read_csv(url, names=column_names, na_values=' ?', skipinitialspace=True)
data = data.dropna()
categorical_columns = ['workclass', 'education', 'marital-status', 'occupation',
                       'relationship', 'race', 'sex', 'native-country', 'income']

encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = encoder.fit_transform(data[col])

X = data.drop('income', axis=1).values
y = data['income'].values

scaler = StandardScaler() # normalize features using StandardScaler
X = scaler.fit_transform(X)

def sigmoid(z): # sigmoid function
    return 1 / (1 + np.exp(-z))

def predict(X, theta): # prediction function
    return sigmoid(np.dot(X, theta))

def cost_function(X, y, theta): # cost function
    m = len(y)
    h = predict(X, theta)
    cost = (-y * np.log(h) - (1 - y) * np.log(1 - h))
    return np.sum(cost) / m

# gradient descent function
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        h = predict(X, theta)
        gradient = np.dot(X.T, (h - y)) / m
        theta -= learning_rate * gradient

        if i % 100 == 0:
            cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

# set term (bias) to X
X = np.c_[np.ones(X.shape[0]), X]

theta_initial = np.zeros(X.shape[1])

learning_rate = 0.01
iterations = 1000

theta_optimal, cost_history = gradient_descent(X, y, theta_initial, learning_rate, iterations) # train the model using gradient descent
print(f"Final cost: {cost_history[-1]}")

def predict_class(X, theta):
    prob = predict(X, theta)
    return (prob >= 0.5).astype(int)

y_pred = predict_class(X, theta_optimal) # predict on the training set

accuracy = np.mean(y_pred == y)
print(f"Training accuracy: {accuracy * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]

data = pd.read_csv(url, names=column_names, na_values=' ?', skipinitialspace=True)

data = data.dropna()

categorical_columns = ['workclass', 'education', 'marital-status', 'occupation',
                       'relationship', 'race', 'sex', 'native-country', 'income']

encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = encoder.fit_transform(data[col])

X = data.drop('income', axis=1).values
y = data['income'].values

scaler = StandardScaler()
X = scaler.fit_transform(X)
X = np.c_[np.ones(X.shape[0]), X]

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(X, theta):
    return sigmoid(np.dot(X, theta))

def cost_function(X, y, theta):
    m = len(y)
    h = predict(X, theta)

    # avoid log(0) and log(1) errors
    h = np.clip(h, 1e-10, 1 - 1e-10)

    cost = (-y * np.log(h) - (1 - y) * np.log(1 - h))
    return np.sum(cost) / m




def batch_gradient_descent(X, y, theta, learning_rate, iterations):# BGD
    m = len(y)
    cost_history = []

    for i in range(iterations):
        h = predict(X, theta)
        gradient = np.dot(X.T, (h - y)) / m
        theta -= learning_rate * gradient

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history


def stochastic_gradient_descent(X, y, theta, learning_rate, iterations): #SGD
    m = len(y)
    cost_history = []

    for i in range(iterations):
        for j in range(m):
            random_index = np.random.randint(m)
            X_i = X[random_index:random_index+1]
            y_i = y[random_index:random_index+1]

            h = predict(X_i, theta)
            gradient = np.dot(X_i.T, (h - y_i))
            theta -= learning_rate * gradient

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

def mini_batch_gradient_descent(X, y, theta, learning_rate, iterations, batch_size): #MBGD
    m = len(y)
    cost_history = []

    for i in range(iterations):
        for j in range(0, m, batch_size):
            X_batch = X[j:j+batch_size]
            y_batch = y[j:j+batch_size]

            h = predict(X_batch, theta)
            gradient = np.dot(X_batch.T, (h - y_batch)) / batch_size
            theta -= learning_rate * gradient

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

theta_initial = np.zeros(X.shape[1])

learning_rate = 0.01
iterations = 1000
batch_size = 32

# train the model using all three variants
theta_bgd, cost_bgd = batch_gradient_descent(X, y, theta_initial, learning_rate, iterations)
theta_sgd, cost_sgd = stochastic_gradient_descent(X, y, theta_initial, learning_rate, iterations)
theta_mbgd, cost_mbgd = mini_batch_gradient_descent(X, y, theta_initial, learning_rate, iterations, batch_size)

plt.figure(figsize=(10, 6))
plt.plot(cost_bgd, label='Batch Gradient Descent')
plt.plot(cost_sgd, label='Stochastic Gradient Descent')
plt.plot(cost_mbgd, label='Mini-batch Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
plt.title('Comparison of Gradient Descent Variants')
plt.show()

print(f"Final cost (BGD): {cost_bgd[-1]}")
print(f"Final cost (SGD): {cost_sgd[-1]}")
print(f"Final cost (MBGD): {cost_mbgd[-1]}")
```

Key Differences Between the Variants:

Batch Gradient Descent (BGD):
- Updates the parameters using entire dataset
- Computationally expensive for large datasets but provides stable smooth convergence
- Stable and smooth convergence, but slow for large datasets
- With longer computation, cost function decreases smoothly  

Stochastic Gradient Descent (SGD):

- Updates parameters using one random training example
- Faster convergence but higher variance and noisy with potential overshooting
- Cost function fluctuates significantly during optimization

Mini-batch Gradient Descent (MBGD):
- Faster convergence with stability, so provides smoother updates compared to SGD.
-  Faster convergence than BGD, smoother than SGD. Cost function decreases more smoothly.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]

data = pd.read_csv(url, names=column_names, na_values=' ?', skipinitialspace=True)

data = data.dropna()

categorical_columns = ['workclass', 'education', 'marital-status', 'occupation',
                       'relationship', 'race', 'sex', 'native-country', 'income']

encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = encoder.fit_transform(data[col])

X = data.drop('income', axis=1).values
y = data['income'].values

scaler = StandardScaler()
X = scaler.fit_transform(X)

X = np.c_[np.ones(X.shape[0]), X]

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(X, theta):
    return sigmoid(np.dot(X, theta))

def cost_function(X, y, theta):
    m = len(y)
    h = predict(X, theta)
    h = np.clip(h, 1e-10, 1 - 1e-10)
    cost = (-y * np.log(h) - (1 - y) * np.log(1 - h))
    return np.sum(cost) / m

def stochastic_gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        for j in range(m):
            random_index = np.random.randint(m)
            X_i = X[random_index:random_index+1]
            y_i = y[random_index:random_index+1]

            h = predict(X_i, theta)
            gradient = np.dot(X_i.T, (h - y_i))
            theta -= learning_rate * gradient

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

# momentum
def momentum_optimizer(X, y, theta, learning_rate, iterations, beta=0.9):
    m = len(y)
    v = np.zeros_like(theta)
    cost_history = []

    for i in range(iterations):
        h = predict(X, theta)
        gradient = np.dot(X.T, (h - y)) / m
        v = beta * v + (1 - beta) * gradient
        theta -= learning_rate * v

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

# RMSProp
def rmsprop_optimizer(X, y, theta, learning_rate, iterations, beta=0.9, epsilon=1e-8):
    m = len(y)
    v = np.zeros_like(theta)
    cost_history = []

    for i in range(iterations):
        h = predict(X, theta)
        gradient = np.dot(X.T, (h - y)) / m
        v = beta * v + (1 - beta) * gradient ** 2
        theta -= learning_rate * gradient / (np.sqrt(v) + epsilon)

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

# adam
def adam_optimizer(X, y, theta, learning_rate, iterations, beta1=0.9, beta2=0.999, epsilon=1e-8):
    m = len(y)
    m_t = np.zeros_like(theta)
    v_t = np.zeros_like(theta)
    cost_history = []

    for i in range(iterations):
        h = predict(X, theta)
        gradient = np.dot(X.T, (h - y)) / m

        m_t = beta1 * m_t + (1 - beta1) * gradient
        v_t = beta2 * v_t + (1 - beta2) * gradient ** 2

        m_t_hat = m_t / (1 - beta1 ** (i + 1))
        v_t_hat = v_t / (1 - beta2 ** (i + 1))

        theta -= learning_rate * m_t_hat / (np.sqrt(v_t_hat) + epsilon)

        cost_history.append(cost_function(X, y, theta))

    return theta, cost_history

learning_rate = 0.01
iterations = 50

# SGD
theta_initial_sgd = np.zeros(X.shape[1])
theta_sgd, cost_sgd = stochastic_gradient_descent(X, y, theta_initial_sgd, learning_rate, iterations)

# Momentum
theta_initial_momentum = np.zeros(X.shape[1])
theta_momentum, cost_momentum = momentum_optimizer(X, y, theta_initial_momentum, learning_rate, iterations)

# RMSProp
theta_initial_rmsprop = np.zeros(X.shape[1])
theta_rmsprop, cost_rmsprop = rmsprop_optimizer(X, y, theta_initial_rmsprop, learning_rate, iterations)

# Adam
theta_initial_adam = np.zeros(X.shape[1])
theta_adam, cost_adam = adam_optimizer(X, y, theta_initial_adam, learning_rate, iterations)

def get_predictions(theta):
    return (predict(X, theta) >= 0.5).astype(int)

y_pred_sgd = get_predictions(theta_sgd)
y_pred_momentum = get_predictions(theta_momentum)
y_pred_rmsprop = get_predictions(theta_rmsprop)
y_pred_adam = get_predictions(theta_adam)

print("SGD Evaluation Metrics:")
print(f"Precision: {precision_score(y, y_pred_sgd)}")
print(f"Recall: {recall_score(y, y_pred_sgd)}")
print(f"F1-Score: {f1_score(y, y_pred_sgd)}")

print("\nMomentum Evaluation Metrics:")
print(f"Precision: {precision_score(y, y_pred_momentum)}")
print(f"Recall: {recall_score(y, y_pred_momentum)}")
print(f"F1-Score: {f1_score(y, y_pred_momentum)}")

print("\nRMSProp Evaluation Metrics:")
print(f"Precision: {precision_score(y, y_pred_rmsprop)}")
print(f"Recall: {recall_score(y, y_pred_rmsprop)}")
print(f"F1-Score: {f1_score(y, y_pred_rmsprop)}")

print("\nAdam Evaluation Metrics:")
print(f"Precision: {precision_score(y, y_pred_adam)}")
print(f"Recall: {recall_score(y, y_pred_adam)}")
print(f"F1-Score: {f1_score(y, y_pred_adam)}")

plt.figure(figsize=(10, 6))
plt.plot(cost_sgd, label='SGD')
plt.plot(cost_momentum, label='Momentum')
plt.plot(cost_rmsprop, label='RMSProp')
plt.plot(cost_adam, label='Adam')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
plt.title('Comparison of Optimization Algorithms')
plt.show()

print(f"Final cost (SGD): {cost_sgd[-1]}")
print(f"Final cost (Momentum): {cost_momentum[-1]}")
print(f"Final cost (RMSProp): {cost_rmsprop[-1]}")
print(f"Final cost (Adam): {cost_adam[-1]}")
```

Each optimization algorithm is unique with its own specific strengths and weaknesses. User can chose based on their needs.

SGD is straightforward and computationally cheap, making it great for large datasets. However, it can take a long time to converge, especially if the learning rate isn't well-tuned, and it struggles with noisy gradients or complex optimization surfaces.

Momentum improves on SGD by smoothing the path to convergence, reducing oscillations, and speeding things up. It does well for problems with lots of local variations but requires careful tuning of the beta parameter to achieve the best result.

RMSProp is designed for problems where the gradient keeps on changing. It dynamically changes the learning rate for every parameter, which makes it more powerful for non-stationary problems than SGD or Momentum. However, the extra hyperparameters like beta and epsilon add to the complexity in tuning.

Adam is the best of both worlds, with Momentum and RMSProp combined. It does both speed and stability in a very small general version, making it often the default choice for many problems since it tends to work well out-of-the box. However, it can be much more computationally intensive and sometimes requires fine-tuning to avoid overfitting or getting stuck in plateaus. In practice, the trade-offs boil down to the problem size, complexity, and resources. For massive datasets, simpler methods like SGD might be more practical due to their low memory needs, while adaptive methods like Adam shine when dealing with tricky optimization surfaces or problems that require fast, stable convergence.
